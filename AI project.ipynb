```python
pattern = r'\b(HR|DESIGNER|INFORMATION TECHNOLOGY|TEACHER|ADVOCATE|BUSINESS DEVELOPMENT|HEALTHCARE|FITNESS|AGRICULTURE|BPO|SALES|CONSULTANT|DIGITAL MEDIA|AUTOMOBILE|CHEF|FINANCE|APPAREL|ENGINEERING|ACCOUNTANT|CONSTRUCTION|PUBLIC RELATIONS|BANKING|ARTS|AVIATION)\b'

df['POSITION_TITLE'] = df['Resume_str'].str.extract(pattern, expand=False)

df['POSITION_TITLE'] = df['POSITION_TITLE'].fillna(method='ffill')
for match in df['POSITION_TITLE']:
    print(match)
```

```python
pattern = r'\b (Ph.D.|phd|master|Masters|bachelor|Bachelors|diploma|high school|secondary|"B.S")\b'
df['EDUCATION _TYPE'] = df['Resume_str'].str.findall(pattern,flags=re.IGNORECASE)
priority_mapping = {'phd': 4, 'master': 3, 'bachelor': 2, 'diploma': 1, 'high school': 0, 'secondary': 0, 'B.S': 2,'Masters':3,'masters':3,'bachelors':2}
mode_EDUCATION_TYPE = df['EDUCATION _TYPE'].mode()
df['EDUCATION _TYPE'] = df['EDUCATION _TYPE'].fillna(mode_EDUCATION_TYPE)#fill EDUCATION _TYPE with mode
for match in df['EDUCATION _TYPE']:
    if match:
        highest_priority = max(match, key=lambda x: priority_mapping[x.lower()])
        print(highest_priority)
```

```python
pattern = r'\b (university|college|high school)\b'
df['SCHOOL_TYPE'] = df['Resume_str'].str.findall(pattern,flags=re.IGNORECASE)
priority_mapping = {'university': 3, 'college': 2, 'high school': 1}
mode_SCHOOL_TYPE = df['SCHOOL_TYPE'].mode()#fill SCHOOL_TYPE with mode
df['SCHOOL_TYPE'] = df['SCHOOL_TYPE'].fillna(mode_SCHOOL_TYPE)
for match in df['SCHOOL_TYPE']:
    if match:
        highest_priority = max(match, key=lambda x: priority_mapping[x.lower()])
        print(highest_priority)
```

```python
pattern = r'\b (aword|awords|Prize|bounty|premium|recompenses|win|accolades|Prizes|reward)\b'
df['AWORDS'] = df['Resume_str'].str.extract(pattern, expand=False,flags=re.IGNORECASE)
df['AWORDS'] = df['AWORDS'].fillna('No')

df.loc[df['AWORDS'] != 'No', 'AWORDS'] = 'Yes'

for match in df['AWORDS']:
    print(match) 
    
```

```python
pattern = r'\b (COMMUNITY SERVICE|voluntary work|volunteer|enlist|recruit|spontaneity|volunteered|unpaid|free of charge )\b'
df['COMMUNITY SERVICE'] = df['Resume_str'].str.extract(pattern, expand=False,flags=re.IGNORECASE)
df['COMMUNITY SERVICE'] = df['COMMUNITY SERVICE'].fillna('No')

df.loc[df['COMMUNITY SERVICE'] != 'No', 'COMMUNITY SERVICE'] = 'Yes'

for match in df['COMMUNITY SERVICE']:
    print(match)        
   
  
```

```python
pattern = r'\b (LICENSE|licence|permit)\b'
df['DRIVERS_LICENSE_AVAILABILITY'] = df['Resume_str'].str.extract(pattern, expand=False,flags=re.IGNORECASE)

df['DRIVERS_LICENSE_AVAILABILITY'] = df['DRIVERS_LICENSE_AVAILABILITY'].fillna('No')

df.loc[df['DRIVERS_LICENSE_AVAILABILITY'] != 'No', 'DRIVERS_LICENSE_AVAILABILITY'] = 'Yes'

for match in df['DRIVERS_LICENSE_AVAILABILITY']:
    print(match)        
   
```

```python
pattern = r'(\d+)(?:\s+)?(?:[+&])?(?:\s+)?(?:year[s]?)'
df['EXPERIENCE_LENGTH'] = df['Resume_str'].str.extract(pattern, expand=False, flags=re.IGNORECASE)

df['EXPERIENCE_LENGTH'] = pd.to_numeric(df['EXPERIENCE_LENGTH'], errors='coerce') # convert column to numeric data
min_length = df['EXPERIENCE_LENGTH'].min() 
df['EXPERIENCE_LENGTH'] = df['EXPERIENCE_LENGTH'].fillna(min_length) # fill null value with  minimum 
df.loc[df['EXPERIENCE_LENGTH'] > 40, 'EXPERIENCE_LENGTH'] = min_length

for match in df['EXPERIENCE_LENGTH']:
    if pd.notnull(match) and int(match) <= 40:
        print(match)
    else:
        print("No Experience")
```

```python
pattern = r'\bGPA:\s+(\d+)'
df['GPA'] = df['Resume_str'].str.extract(pattern, expand=False, flags=re.IGNORECASE)

pattern = r'\bGPA:\s+(\d+)'
df['GPA'] = df['Resume_str'].str.extract(pattern, expand=False, flags=re.IGNORECASE)
df['GPA'] = pd.to_numeric(df['GPA'], errors='coerce') # convert column to numeric data
# mean_length = df['GPA'].mean() 
df['GPA'] = df['GPA'].fillna(2.8)
df.loc[df['GPA'] > 4, 'GPA'] = 2.8

for match in df['GPA']:
    if pd.notnull(match) and int(match) <= 4 :
            print(match)
    else:
        print("no")
```

```python
def extract_words_with_commas(text, keyword):
    pattern = fr"\b{re.escape(keyword)}\b\s*([^.]*)"
    match = re.search(pattern, text)
    if match:
        extracted_text = match.group(1)
        extracted_words = re.findall(r'\b\w+(?:,\s*\w+)+\b', extracted_text)
        if extracted_words:
            return extracted_words
    return None


df['Skills'] = df['Resume_str'].apply(lambda x: extract_words_with_commas(x, keyword="Skills"))

for skills in df['Skills']:
    if skills:
        print(skills)
    else:
        print("Condition not met.")


```

```python
def count_the_num_of_skills(skills_list):
    if skills_list:
        return len(skills_list)
    return 0

df['SKILLS_COUNT'] = df['Skills'].apply(lambda x: count_the_num_of_skills(x))

mean_skills_count = df['SKILLS_COUNT'].mean()
df['SKILLS_COUNT'] = df['SKILLS_COUNT'].fillna(mean_skills_count) # fill null values with mean


for count in df['SKILLS_COUNT']:
    if pd.notnull(count) and int(count) >0:
        print(count)        
    else:
        print("There are no skills.")
```

### <span style="color:brown">Q2:

A concerned parents contacts you to tell them if there is a specific school type that would guaranteed for their child a good future if you can tell, and which is it ?

```python
from collections import Counter
school_types = [item.lower() for sublist in df['SCHOOL_TYPE'] for item in sublist]
counts = Counter(school_types)

for school_type, count in counts.items():
    print(f"{school_type}: {count}")
```

### <span style="color:brown">Q3:

what is the most common skills among all applications?

```python
from collections import Counter
all_skills = []
for skills in df['Skills']:
    if skills:
        all_skills.extend(skills)

skills_counter = Counter(all_skills)

most_common_skills = skills_counter.most_common()

for skill, frequency in most_common_skills:
    print(f"Skill: {skill}, Frequency: {frequency}")
```

### <span style="color:brown">Q4:

Based on the data that you have, could you tell the most active and community helpful applicants are for what position?

```python
position_counts = {}

for index, row in df.iterrows():
    position = row['POSITION_TITLE']
    community_service = row['COMMUNITY SERVICE']

    if pd.notna(community_service):
        position_counts[position] = position_counts.get(position, 0) + 1

if position_counts:
    most_community_helpful_position = max(position_counts, key=position_counts.get)
else:
    most_community_helpful_position = "No community-helpful applicants found"

print("Most Community Helpful Position:", most_community_helpful_position)
```

### <span style="color:brown">Q5:

could you nominate the best job application for each position?

```python
best_applications = df.groupby('POSITION_TITLE')['EXPERIENCE_LENGTH'].idxmax()
best_job_applications = df.loc[best_applications]

for index, row in best_job_applications.iterrows():
    position = row['POSITION_TITLE']
    experience_length = row['EXPERIENCE_LENGTH']
    
    if pd.notnull(experience_length) and int(experience_length) <= 40:
        print(f"Best application for {position} position: {experience_length} years of experience")
    else:
        print(f"No experience found for {position} position")
```

```python
max_experience_ids = df.groupby('POSITION_TITLE')['EXPERIENCE_LENGTH'].idxmax() # The ID for the maximum experience position

for position, index in max_experience_ids.items():
    experience_length = df.loc[index, 'EXPERIENCE_LENGTH']
    if pd.notnull(experience_length) and int(experience_length) <= 40:
        print(f"Position: {position}\tMax Experience ID: {df.loc[index, 'ID']}\tExperience Length: {experience_length} years")
    else:
        print(f"No experience found for {position} position")
```
